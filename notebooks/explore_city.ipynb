```python
import pickle
import matplotlib.pyplot as plt

# Load extracted data
with open('data/processed/sjc.pkl', 'rb') as f:
    state = pickle.load(f)

print(state.summary())

# Visualize
fig, axes = plt.subplots(1, 2, figsize=(16, 8))

# Left: Streets only
state.streets.plot(ax=axes[0], color='gray', linewidth=0.3)
axes[0].set_title('Streets Network\n(37,044 segments)')
axes[0].axis('off')

# Right: Streets + Blocks
state.streets.plot(ax=axes[1], color='gray', linewidth=0.3, alpha=0.5)
state.blocks.plot(ax=axes[1], facecolor='lightblue', edgecolor='blue', 
                   alpha=0.6, linewidth=0.2)
axes[1].set_title('Streets + Blocks\n(6,852 blocks)')

axes[1].axis('off')
plt.tight_layout()
plt.savefig('data/processed/sjc_extraction.png', dpi=150, bbox_inches='tight')
plt.show()

print("\nâœ“ Visualization saved to data/processed/sjc_extraction.png")
```
